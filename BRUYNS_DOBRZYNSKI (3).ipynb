```python
import numpy as np
import matplotlib.pyplot as plt
```

## Retrieving the data
First, we retrieve the data that we will study.

```python
data_lang = np.genfromtxt('Vpp=20_10Pa.csv', delimiter = '","', skip_header=1)
```

```python
# C2 Data
v_wrong = data_lang[:, 1] 
current_wrong = data_lang[:, 2] 

def extract_values(v_right, v_wrong, current_right, current_wrong, start, end):
    for k in range(start,end) :
        if v_wrong[k-1] < v_wrong[k] :
            v_right.extend([v_wrong[k]])
            current_right.extend([current_wrong[k]])
            

v_right = []
current_right = []
extract_values(v_right, v_wrong, current_right , current_wrong, 19, 2010)


# print(v_right)
# Create the graphic
plt.plot( v_right, np.log(current_right), '.', label='noisy', color='blue')
plt.plot( v_right, current_right, '.', label='noisy', color='red')

# Add labels and a title
plt.ylabel('V')
plt.title('C2 Data')

# Show the legend
plt.legend()

# Show the graphic
plt.show()
```
